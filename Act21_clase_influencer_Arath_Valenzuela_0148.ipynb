```python
class Influencer:
    # El constructor define los atributos
    def __init__(self, nombre, red_social, seguidores):
        self.nombre = nombre
        self.red_social = red_social
        self.seguidores = seguidores

    # Función 1: Publicar
    def publicar(self):
        print(f"📸 {self.nombre} ha subido un nuevo post en {self.red_social}.")

    # Función 2: Colaborar
    def colaborar(self):
        print(f"🤝 {self.nombre} está realizando una colaboración pagada.")

    # Función 3: Interactuar
    def interactuar_con_seguidores(self):
        print(f"💬 {self.nombre} está respondiendo comentarios a sus {self.seguidores} seguidores.")

# --- Creación de Objetos (Instancias) ---

# influencer1: Datos de ejemplo
influencer1 = Influencer("Santi Tech", "Instagram", "50k")

# influencer2: Datos de ejemplo
influencer2 = Influencer("Mari GaminG", "Twitch", "1.2M")

# --- Ejecución de acciones ---

print("--- Actividad de Influencer 1 ---")
print(f"Nombre: {influencer1.nombre}")
influencer1.publicar()
influencer1.colaborar()
influencer1.interactuar_con_seguidores()

print("\n--- Actividad de Influencer 2 ---")
print(f"Nombre: {influencer2.nombre}")
influencer2.publicar()
influencer2.colaborar()
influencer2.interactuar_con_seguidores()
print("-------------------------------------------------------------------")
print(" Programa creado por Arath Valenzuela (NL52) en ayuda de Gemini AI")
```
